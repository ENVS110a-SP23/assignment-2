## Question 3: Data structures (5 pt)

- Create a dictionary with 3 keys called `'a'`, `'b'`, and `'c'`. The value for each key should be a blank list. (1 pt)
- Append the list associated with key `a` with a blank set. (1 pt)
- Append the list associated with key `b` with any 4 numbers. (1 pt)
- Change the third item in the list associated with key `b` to be `1000`. (1 pt)
- Print out the dictionary (1 pt)

```python
# your code here:

new_dict = {
    'a' : [],
    'b' : [],
    'c' : []}

new_dict['a'].append(set())
new_dict['b'].append(1)
new_dict['b'].append(2)
new_dict['b'].append(3)
new_dict['b'].append(4)

new_dict['b'][2] = 1000
print(new_dict)


```

```python
# don't modify the lists

temperatures = [29.75, 12.63, 31.58, 7.16, 32.51]

populations = [442, 336, 505, 913, 933]

dates = ['5/25/2022','3/18/2022','6/28/2022','11/11/2022','7/6/2023']

# your code below:

data_dict = {
    'Haley McCann' : [],
    'Siena Welch' : [],
    'Jaylin Mercado' : [],
    'Ismael Hayden' : [],
    'Nina Bright' : []
}

Haley_set = set()
Haley_set = {temperatures[0], populations[0], dates[0]}
Siena_set = set()
Siena_set = {temperatures[1], populations[1], dates[1]}
Jaylin_set = set()
Jaylin_set = {temperatures[2], populations[2], dates[2]}
Ismael_set = set()
Ismael_set = {temperatures[3], populations[3], dates[3]}
Nina_set = set()
Nina_set = {temperatures[4], populations[4], dates[4]}


data_dict['Haley McCann'].append(Haley_set)
data_dict['Siena Welch'].append(Siena_set)
data_dict['Jaylin Mercado'].append(Jaylin_set)
data_dict['Ismael Hayden'].append(Ismael_set)
data_dict['Nina Bright'].append(Nina_set)

print(data_dict)

```
